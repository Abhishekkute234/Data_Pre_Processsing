# Function Transformation 

### convert non-distribution data to distribution data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_1=pd.read_csv("Online Sales Data.csv")
df_1.head(5)

# here UNIT PRICE and TOTAL REVENUE are the two non distributional data column
```

```python
df_1.isnull().sum()
```

```python
df_1["Unit Price"].fillna(df["Unit Price"].mode()[0],inplace=True)
df_1["Total Revenue"].fillna(df["Total Revenue"].mode()[0],inplace=True)
df_1.isnull().sum()
```

```python
sns.displot(df_1["Total Revenue"])
plt.show()
```

```python
q1=df_1["Total Revenue"].quantile(0.25)
q3=df_1["Total Revenue"].quantile(0.75)

IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q1+(1.5*IQR)
min_range,max_range
```

```python
df_1=df_1[df_1["Total Revenue"]<=max_range]
```

```python
sns.displot(df_1["Total Revenue"])
plt.show()
```

```python
# so this is non normal distribution data

from sklearn.preprocessing import FunctionTransformer
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(df_1[["Total Revenue"]])
```

```python
df_1["tran_Total Revenue"]=ft.transform(df_1[["Total Revenue"]])
df_1
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.displot(df_1["Total Revenue"])
plt.title("Before")

plt.subplot(1,2,2)
sns.displot(df_1["tran_Total Revenue"])
plt.title("After")


plt.show()
```
