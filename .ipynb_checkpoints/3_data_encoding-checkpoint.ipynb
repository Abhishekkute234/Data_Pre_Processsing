# Data Encoding (one hot Coding)

# This is for ordinal data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

dataset=pd.read_csv("Dataset salary 2024.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
for i in dataset.columns:
   dataset[i].fillna(dataset[i].mode()[0],inplace=True)
dataset.isnull().sum()

```

# No.1 Get dummies using pandas

```python
en_data=dataset[["experience_level","employment_type"]]
en_data
```

```python
pd.get_dummies(en_data)
# it convert it to boolean values
```

# Data Encoding using Sk learn 

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()
ar=ohe.fit_transform(en_data).toarray()
ar
```

```python
data = ar  # Replace `ar` with your actual data

# Define column names (ensure the number matches the number of columns in `data`)
columns = ["experience_level_EN", "experience_level_EX",
           "experience_level_MI", "experience_level_SE",
           "employment_type_CT", "employment_type_FL",
           "employment_type_FT", "employment_type_PT",
          ]

# Create the DataFrame
try:
    df = pd.DataFrame(data, columns=columns)
    print(df)  # Print the DataFrame for verification
except ValueError as e:
    if "Shape of passed values" in str(e):  # Check for shape mismatch error
        print("Error: The number of columns in your data doesn't match the provided column names.")
        print("Please adjust the column names or verify the data structure.")
    else:
        raise e  # Re-raise other unexpected errors
```
