# Label Encoding

#### This is for nominal data( no connections)

```python
import pandas as pd
df=pd.DataFrame({"name":["lion","dog","cat","cow","black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["en_name"]=le.fit_transform(df["name"])
df
```

```python
dataset=pd.read_csv("Online Sales Data.csv")
dataset
```

```python
lab=LabelEncoder()
lab.fit(dataset["Product Name"])
```

```python
dataset["Product Name"]=lab.transform(dataset["Product Name"])
```
