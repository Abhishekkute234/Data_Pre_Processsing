# Handiling missing values using sklearn

```python
from sklearn.impute import SimpleImputer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dataset=pd.read_csv("Crime_2015.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info() 
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
si= SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[['Murder', 'Rape', 'AggravatedAssault', 'MotorVehicleTheft']])
```

```python
new_dataset=pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
new_dataset
```

```python
new_dataset.isnull().sum()
```
