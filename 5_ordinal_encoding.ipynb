# Ordinal Encoding(1st method)

```python
import pandas as pd

data = ["s", "m", "l", "xl", "l", "xll", "m", "s", "xl", "l"]  # List of data
df = pd.DataFrame(data, columns=["Size"])  # Create DataFrame with named column
print(df)
```

```python
# give the order
ord_data=[["s","m","l","xl","xll"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["en_size"]=oe.transform(df[["Size"]])
df
```
